```python
import re
import pandas as pd
from datetime import date
```

```python
df = pd.read_excel("./cleaned/cleaned_2023-09-10.xlsx", dtype="str")

df.fillna("", inplace=True)

df.loc[df["PREVIOUS  OWNER"].str.contains(r"HercRentals", flags=re.I, regex=True), 
       "PREVIOUS  OWNER"] = "Herc Rentals"

df.head()
```

```python
df_products = df.to_dict("records")

df_products[:3]
```

```python
df_res = pd.read_excel("./data/results_2023-09-08  COMPLETED.xlsx", dtype="str")

df_res.fillna("", inplace=True)

df.head()
```

```python
df_res_products = df_res.to_dict("records")

df_res_products
```

```python
for item in df_products:
    owner: str = item["PREVIOUS  OWNER"]
    
    item["PREVIOUS  OWNER"] = owner.upper()

for old_item in df_products[2700:]:
    previous_item = (old_item["SALE DATE"], 
                     old_item["BIDADOO PRICE"],
                     old_item["LINK TO LISTING"])
    
    for new_item in df_res_products[2700:]:
        if previous_item == (new_item["SALE DATE"],
                             new_item["BIDADOO PRICE"],
                             new_item["LINK TO LISTING"]):
            owner: str = new_item["PREVIOUS  OWNER"]

            if owner.strip():
                old_item["PREVIOUS  OWNER"] = owner.upper()

            break
```
